## Introduction

Context: The LibraryReads lists capture the top 10 newly published books library employees and volunteers are most excited to recommend each month. A place on the list is based on a popular vote, not on an explicit evaluation of literary merit, and winners receive a considerable amount of publicity. Five years after the list was created, a "Hall of Fame" policy was established: once an author had two works appear on LibraryReads Lists, any subsequent books that reached the top 10 by vote count would receive marketing as a book by a Hall of Fame author, and would not count towards the top 10 books for the month. The policy was instituted to increase the variety of content in the LibraryReads list. 

Research question: Did the creation of the LibraryReads Hall of Fame actually increase variety in LibraryReads Top 10 lists? In this project variety will be measured by number of unique authors, presence of debut authors on the list, and by genre category. 

Secondary questions: How successfully can genre/form information from Library of Congress records be used to classify LibraryReads books into more general genre categories? Can information from the annotations provided in LibraryReads marketing data used to be identify genres of books?

Summary:
The introduction of the LibraryReads Hall of Fame is associated with a larger number of unique authors being featured in at least one list over the course of a year. This variation does not stem from the inclusion of debut authors, who continue to be included at an average rate of approximately 2 debut authors per list. Using Library of Congress catalog records, a multinomial logit model is able to correctly predict the assigned LibraryReads genre in approximately 42% of cases, however the model is only able to classify works in half of the genres that exist within the LibraryReads archive. When using the predicted genre to compare the pre-Hall of Fame period and the Hall of Fame period, there is no evidence of a statistically significant change in distribution of genres within a list. Using the text of the annotations to predict genre is less successful, resulting in classifications in only two of the nine genres represented among LibraryReads annotations.

## Data Description

### Context of Raw Datasets
LibraryReads is a nonprofit organization that invites individuals affiliated with public libraries in the United States to vote for their favorite newly published book in an upcoming month. The top 10 vote receiving new publications for adult readers in a given month are published as the LibraryReads List, with substantial promotional materials available for marketing the books. Votes are collected via the Edelweiss and NetGalley platforms, both of which are hubs for publishers to make ARC (advanced reading copies) of books available to library staff, booksellers, and other members of the book reviewing community. LibraryReads defines an individual as affiliated with a public library eligible to vote for titles when they are a staff member or a volunteer who list the library they are associated with in their Edelweiss/NetGalley profile (https://libraryreads.org/for-library-staff). Publishers control access to ARCs via permissions on the distribution platforms, and/or by selectively offering print copies of pre-publication works.

Support for the LibraryReads organization and the publication of the lists comes from the Edelweiss and NetGalley platforms, publications that print the list, the Adult Library Marketing Association, and various publishers (LibraryReads Annual Report, 2020). Publishers are solicited for funding support after one of their books has been included in a LibraryReads List (LibraryReads FAQ). 

Each observation is a work of fiction or non-fiction for adult audiences, voted by library workers and volunteers as one of the new publications they are most excited to share with patrons that month. For each work, the dataset includes the year and month of inclusion on the LibraryReads List, author, title, publication date, publisher, and 13-digit ISBN. Indicators are included to denote the top ranked book for each month, hall of fame authors (a program started in October 2018, where any author with 3 or more appearances on any LibraryReads list no longer takes one of the top 10 positions but is included in marketing materials), debut authors, and end of the year favorites.   

Supplementing the LibraryReads data is information from the Library of Congress Catalog, focusing on the form/genre field of the record. The catalog is publically accessible, and records follow the standardized MARC format. Most, but not all, publications are incorporated into the catalog records of the Library of Congress. 

The creation and publication of the LibraryReads Archive is a relatively new endeavour by the LibraryReads organization (LibraryReads Annual Report, 2020). The information has been available in the form of the monthly PDF and promotional tools, but the conversion to a spreadsheet format for easy use and filtering is a gradual process. For this reason, some fields are not recorded for all records. In particular, the genre, nominator, and recommendation text are not available for all observations. The Library of Congress is a prominent source of MARC records for libraries, and has holdings or records for most publications, but not all publications merit inclusion in the publically available Library of Congress Catalog. Form/genre is not a required field in MARC records, and the individual who created the catalog record may have chosen not to include the information in the entry.   

### Preprocessing

The creation and maintenance of the LibraryReads archive is not without error or inconsistencies. Specific details can be found in Appendix 1: Data Cleaning Description. Modifications were made to ensure that each observation had one and only one thirteen digit ISBN associated with the record, to be used as the key field for merging LibraryReads and Library of Congress data. Publication date and genre were standardized to remove format and spelling errors. Publisher names were consolidated to their most common presentation in the data, after verifying that the different names did not represent separate imprints or branches within a larger publishing house. Indicators for debut authors (first published novel), Hall of Fame authors (3 or more appearances on LibraryReads lists), the top pick for the month, and if the book was selected as a "best of the year" title were recoded as binary variables. 

In some cases, the data scraped from the Library of Congress catalog presented the ISBN-10 first rather than the ISBN-13, resulting in a no-match outcome. These ISBNs were replaced with the appropriate ISBN-13 for the edition featured in the LibraryReads list. Duplicate entries of a single genre or form within a record were removed for clarity, and genre/form name fields were consolidated. Due to a high level of differentiation within genre/form headings, many headings were associated with only one record. Consolidation primarily included combining categories of type "*genre descriptor* stories" and "*genre descriptor* fiction" and correcting spelling errors. The name of the combined category follows the more commonly appearing genre/form category name within the data.  

Two formats of the data were produced for analysis: one, as described above, is a single observation for each work included on a LibraryReads List. Genre information from the Library of Congress is recorded across multiple variables for each observation. The second format has each unique ISBN-Library of Congress genre pair as an observation, referred to as the "long dataset." This allows for counts of the number of times any one form/genre heading appears in the data, regardless of what other form/genre headings are also associated with the ISBN.

## Preregistration Statement

### Analysis 1

The Hall of Fame Authors section of the LibraryReads List was created in October 2018 to generate more variety in the LibraryReads Lists. To assess the efficacy of this policy, I will run a paired t-test comparing the average proportion of debut novels on a single list prior to the creation of the Hall of Fame and after the creation of the Hall of Fame. I will also conduct a paired t-test comparing the total proportion of books by debut authors in the Hall of Fame era to the proportion of debut authors in the same number of lists prior to the institution of the Hall of Fame. To account for potential cyclical variation in the months books by debut authors are released, a third variation on the t-test will compare the proportion of debut authors in lists from October 2015 - March 2018 to October 2018 - March 2021. 

Hall of Fame Period: October 2018 - March 2021  
Pre-Hall of Fame Period 1: April 2016 - September 2018  
Pre-Hall of Fame Period 2: October 2015 - March 2018

*Note: The wording of the pre-registered analysis suggests that I intended to run t-test on cases where sample size = 1 (total proportion of books by debut authors in each period). Using all LibraryReads Lists published prior to the Hall of Fame period, I simulated the distribution of possible mean debut authors per list over 29 lists, the number of lists published in the Hall of Fame period. The observed mean debut authors for the Hall of Fame period is within the 90% confidence interval for the mean using the pre-Hall of Fame data, indicating that the mean is within the likely range of outcomes for patterns in the pre-Hall of Fame data. 

```python
#simulation
non_hof_pd = monthly_agg.loc[monthly_agg.hof_pd == 0]
bootstrap_means = np.zeros(10000)
bootstrap_max = np.zeros(10000)
for i in range(10000):
    iteration = non_hof_pd.sample(n=len(monthly_agg.loc[monthly_agg.hof_pd ==1]),replace=True)
    bootstrap_means[i] = iteration.debut.mean()
    bootstrap_max[i]= iteration.debut.max()

#99% confidence interval for mean
conf_int = np.quantile(bootstrap_means,q=[0.005,0.995])
lower_bound = conf_int[0]
upper_bound=conf_int[1]

#90% confidence interval for mean
conf_int2=np.quantile(bootstrap_means,q=[0.05,0.95])
lower_bound2=conf_int2[0]
upper_bound2=conf_int2[1]
    
#plot
plt.figure(figsize=(10,4))
plt.hist(bootstrap_means,bins=int(bootstrap_max.max())+1,color='skyblue')
plt.xlabel('Mean debuts per list')
plt.title('Bootstrapped Distribution of mean debuts per list for 29 lists')
plt.vlines(x=lower_bound, ymin=0,ymax=4000,color='gray',label='99% Confidence Interval bounds')
plt.vlines(x=upper_bound,ymin=0,ymax=4000,color='gray')
plt.vlines(x=lower_bound2, ymin=0,ymax=4000,color='black',label = '90% Confidence Interval bounds')
plt.vlines(x=upper_bound2,ymin=0,ymax=4000,color='black')
plt.vlines(x=hof_pd_avg_debut,ymin=0,ymax=4000,color='red',label='Observed Hall of Fame mean')
plt.legend()
plt.show()

print('99% confidence interval: [{:.2f}, {:.2f}]'.format(lower_bound,upper_bound))
print('90% confidence interval: [{:.2f}, {:.2f}]'.format(lower_bound2,upper_bound2))
print('Observed mean, Hall of Fame Period: {:.2f}'.format(hof_pd_avg_debut))
```

#### Variety in Genre

The stated purpose of the Hall of Fame is to increase variety. With an increase in the number of unique authors, do we observe a substitution for authors in the same genres, or do the new authors represent an increase in the number or prevalence of other genres? 

LibraryReads assigned genre is available for 693 of the 936 books in the archive, and contains both form (short stories, graphic novels) and genre (mystery, literary fiction) classifications. No genres are assigned for books appearing in lists after August 2019. To assess the impact of Hall of Fame on the genre books in the Top 10 lists, it is necessary to provide genre labels for all works. To assign genre for cases where no LibraryReads genre is provided in the archive, data from the Library of Congress catalog records will be used to predict LibraryReads genre. Library of Congress form/genre attributes are available for 649 of the records. 

72 distinct genre/form attributes were identified in the Library of Congress data; however, using all 72 genre/form attributes would likely lead to overfitting the data. OneHotEncoder was used on the long form of the dataset to create 72 dummy variables, one for each individual genre/form attribute. 

The benefit of this approach is the removal of any hierarchical structure assumed from the order of genre/form attributes in the Library of Congress Data. Applying PCA to the wide format would likely result in heavy loading on the variables containing the first two genre/form attributes, and then rapidly decline based on the low number of observations with 3 or more genres. As sorting of the genre/feature attributes is arbitrary in this analysis, it cannot be assumed that the first genre is the most descriptive for any individual work. PCA was chosen to reduce those 72 attributes to 8 variables, which were used in a multinomial logistic regression to predict class belonging based on the the PCA genre/form variables. Other possibilities involving interactions between genre/form attributes are discussed below in the possible extensions for research section. 

As a genre, Literary Fiction dominates the entirety of the LibraryReads archive. As a genre, the definition is relatively fluid. *Readers' Advisory Service in the Public Library* (Saricks, 2005) offers the concise explanation that literary fiction is "\[c\]ritically acclaimed, often award-winning, character-centered, provocative, elegantly written works" (182). The tasks of defining what it means to be literary and developing a more precise definition for the genre are outside the scope of this project, but is important to note that the term may act as a catch-all.

Individual genre and form classifications are coded as indicator variables using scikit-learn's OneHotEncoder. Using primary component analysis, the 71 distinct genre and form categories are reduced to eight principle components. The principle components have an explained variance ratio of at least 0.05. 

```python
#Generate indicator variables
onehot_encoder_genre=OneHotEncoder(sparse = False)
onehot_data = onehot_encoder_genre.fit_transform(long_df[['value']])
onehot_data.shape
column_names=[]
for count, value in enumerate(onehot_encoder_genre.categories_[0]):
    column_names.append(value)

onehot_df_genre = pd.DataFrame(onehot_data,columns=column_names)
onehot_df_genre = pd.merge(long_df,onehot_df_genre,how='left',left_index=True,right_index=True)

#PCA
pca = PCA(n_components=8)
pca_genre = pca.fit_transform(onehot_df_genre.loc[:,'adventure_fiction':])
long_df[['pca_1','pca_2','pca_3','pca_4','pca_5','pca_6','pca_7','pca_8']]=pca_genre[:,:8]


for i in range(8):
    print('The explained variance ratio of primary component',
          i+1, 'is {:.4f}'.format(pca.explained_variance_ratio_[i]))
```

Using the PCA variables in the long format of the data as independent variables, I predicted the LibraryReads genre for each work with a multinomial logistic regression. In the long format of the data (one row per work and single Library of Congress genre/form classification pair), a single work can be associated with up to 4 distinct predicted LibraryReads genres. The scores for the model, using a train-test split with 20% of the works assigned a LibraryReads genre used as the test group, are printed below. A success in this case should be interpreted as the correct LibraryReads genre is predicted using the principle components derived from the given work-Library of Congress pair. This means that a single work may be associated with multiple success and or failure outcomes. 

```python
lr_genre_list=[]
for i in range(12):
    lr_genre_list.append(long_df.genre.value_counts().index[i])

genre_map={}
for j in range(12):
    genre_map[lr_genre_list[j]] = 1 + j

long_df['genre']=long_df['genre'].map(genre_map)
long_df_exclude = long_df.loc[long_df.genre.isna()].copy() #no LibraryReads assigned genre
long_df_include = long_df.loc[long_df.genre.notna()].copy() #has assigned LibraryReads genre

#Train-test split
x_train,x_test,y_train,y_test = train_test_split(long_df_include.loc[:,'pca_1':],long_df_include.genre,test_size=0.2) 

mult_logit = LogisticRegression()
mult_logit.fit(x_train,y_train)
test_predict = mult_logit.predict(x_test)
train_score = mult_logit.score(x_train,y_train)
test_score = mult_logit.score(x_test,y_test)

cv_scores = cross_val_score(LogisticRegression(),long_df_include.loc[:,'pca_1':'pca_8'],long_df_include['genre'],cv=4)

mean_cv_score = cv_scores.mean()
naive_score = len(long_df_include.loc[long_df_include.genre==1])/len(long_df_include)


long_df_exclude['predicted_genre'] = mult_logit.predict(long_df_exclude.loc[:,'pca_1':])
long_df_include['predicted_genre'] = mult_logit.predict(long_df_include.loc[:,'pca_1':])

predicted_genre_df = long_df_include[['isbn','predicted_genre']]
predicted_genre_df = predicted_genre_df.append(long_df_exclude[['isbn','predicted_genre']])

print('The score for the multinomial logistic regression model using the test data to predict', 
      'the LibraryReads genre is {:.4f}.'.format(test_score),
     'The score for the multinomial logistic regression model using the training data to predict', 
      'the LibraryReads genre is {:.4f}.'.format(train_score))
print('The mean cross-validation score with 4 folds over all works with an assigned LibraryReads genre is {:.4f}'.format(mean_cv_score))
print('The naive score, where literary fiction is assigned as the predicted genre for all works with an assigned LibraryReads genre, is {:.4f}'.format(naive_score))
```

For results that are comparable with the observed behavior of LibraryReads genre (one assignment per work), in cases where there was more than one genre assignment for a work as an outcome for the regression, the modal genre was chosen. If there was no single modal genre for the work, a candidate modal genre for the work was randomly selected as the single predicted genre. The model does not assign all observed LibraryReads genres, only historical fiction, SF/fantasy, romance, mystery, suspense/thriller, and literary fiction are assigned outcomes for the predicted genre.  

```python
modal_genre_dict = {}
for j in predicted_genre_df.isbn.unique():
    modal_genre_dict[j] = np.random.choice(predicted_genre_df.loc[predicted_genre_df.isbn==j].predicted_genre.mode())

long_df['modal_genre']=long_df['isbn'].map(modal_genre_dict)
wide_df['modal_genre']=wide_df['isbn'].map(modal_genre_dict)

#map back to strings for predicted genre (easier to read)
genre_map_undo = {}
for i in range(len(genre_map)):
    genre_map_undo[list(genre_map.values())[i]]=list(genre_map.keys())[i]

wide_df['modal_genre_descr'] = wide_df['modal_genre'].map(genre_map_undo)

#accuracy for modal genre
modal_accuracy = len(wide_df.loc[wide_df.genre==wide_df.modal_genre_descr])/len(wide_df.loc[wide_df.genre.notna()])
naive_modal_accuracy = len(wide_df.loc[wide_df.genre=='Literary Fiction'])/len(wide_df.loc[wide_df.genre.notna()])

print('The accuracy of the model and subsequent selection of the modal genre is {:.4f}.'.format(modal_accuracy), 
     'when examining each work with an assigned LibraryReads genre. The naive assignment of literary fiction',
      'as the modal genre for all works would have an accuracy of {:.4f}'.format(naive_modal_accuracy))
```

To bootstrap confidence intervals, samples of 10 books were randomly selected with replacement from the set of books in the pre-Hall of Fame period (October 2013 - September 2018), and the proportion of each list attributed to the predicted genres was calculated. 

```python
modal_genre_list= wide_df.modal_genre_descr.value_counts().index
iterations = 1000
iteration_df = pd.DataFrame(np.zeros((iterations,5)), columns=modal_genre_list)
for j in range(iterations):
    iteration = wide_df.loc[wide_df.hof_pd == 0].sample(10,replace=True)
    for i in modal_genre_list:
        iteration_df.loc[j,i] = iteration.loc[iteration.modal_genre_descr==i].modal_genre_descr.count()/10

r=0
modal_genre_sim_stats=[[],[],[],[],[],[]]
for k in modal_genre_list:
    lower_bound_genre = np.quantile(iteration_df[k],q=[0.025,0.975])[0]
    upper_bound_genre = np.quantile(iteration_df[k],q=[0.025,0.975])[1]
    genre_mean = iteration_df[k].mean()
    print('Mean proportion of',k,'in simulated lists = {:.2f}, 95% confidence interval = ({:.2f},{:.2f})'.format(genre_mean,lower_bound_genre,upper_bound_genre))
    modal_genre_sim_stats[r]=[k,genre_mean,lower_bound_genre,upper_bound_genre]
    r = r+1
```

### Evaluation of Significance

The average number of debut authors in a single LibraryReads Top 10 list did not change in a statistically significant way with the introduction of the LibraryReads Hall of Fame. For both comparison periods, it is not possible to reject the null hypothesis that the mean in the comparison and the mean in Hall of Fame period are different (p = 0.68 and p = 0.89). Testing for a difference between the average number of debut authors on the Hall of Fame list and all LibraryReads lists does not result in a statistically significant result. The observed mean (2.24 debut authors per list) is within the 90% (1.69-2.41) and 99% (1.48 - 2.66) confidence intervals, which suggests that the observed mean could plausibly be from the same data-generating process as the pre-Hall of Fame distribution. 

Using the principal components generated from the Library of Congress genre/form predictors to classify the genre, the model accurately predicts slightly more than 40% of genres. The naive classification, assigning the most common genre, Literary Fiction, to all books in the archive, results in an accurate classification in 26% of cases. 

The observed mean proportion of a list attributable to a specific genre is within the 95% confidence interval produced from running simulations using the pre-Hall of Fame period works as a sample. This suggests that there was not an underlying change in the data generating process related to genres. 

In comparison with using the Library of Congress genre/form classifiers, using the content of the annotations to predict genre is marginally less accurate (38% accuracy). The results of using the annotation text to classify genres into the designated LibraryReads genre categories reduces the number of labels used to two: literary fiction and historical fiction. Using k-means to develop clustering based on the text content for a range of cluster sizes resulted in silhouette scores close to 0, indicating substantial overlap between the clusters. 

The ANOVA test indicates that there is a difference in average length across genres/lack of genre (p-value = 0.013). The Tukey HSD pairwise comparison does not find a significant difference in average length between any two genres, which may be due to small sample sizes and a relatively large number of pairwise comparisons. 

### Interpretation and Conclusions  

#### Variety in Authors

In looking at the number of unique authors featured in lists by year, it is clear that the creation of the Hall of Fame did allow for entry of new authors to the Top 10 lists. The t-tests demonstrate that these new authors did not come from the pool of debut authors, as there is no statistically significant change in the number of debut authors included in an average list with the creation of the Hall of Fame Lists. Taken together, these observations may suggest that inclusion in the LibraryReads Top 10 lists is at least in part a function of having an existing fan base that is willing to read and/or vote for new books by a given author. To more fully explore this hypothesis, more information about the publishing histories of the authors who appear in the Top 10 lists and in the Hall of Fame would be required. 

#### Variety in Genres

Based on genre classifications stemming from the Library of Congress information, there is no discernable change
in the distribution of genres within a single LibraryReads Top 10 list. This may represent a stable distribution of leisure reading interests and genres recommended by library staff or volunteers. It also may be an artifact of the broad genre categories used to classify works in the LibraryReads Archive, which could mask more subtle variation in trends. 

#### Text Analysis of Annotations

The small sample size of the dataset when restricted to those works with an annotation in the LibraryReads Archive is a likely limitation of this analysis. As noted above, the silhouette scores for clustering using the k-means algorithm with 2 to 9 centroids has a maximum at 0.001, suggesting substantial overlap between clusters. The words that convey the highest amount of mutual information appear to be fairly generic, though they do not appear in more than 15% of any of the annotations. Taken together, this suggests that annotations for LibraryReads recommendations have similar features regardless of genre. Using the controlled vocabulary of shared words allowed the classifier to be applied to the data without an assigned LibraryReads genre, but it is likely that the same vocabulary is by force limited to more common words. 

It is not clear that genre is related to annotation length, which may be in part related to the fact that annotations may be edited by the LibraryReads team for length and clarity. 

#### Possible Extensions of Research
The Hall of Fame provides data to examine a counter factual scenario: books are only mentioned in the Hall of Fame when they woud normally merit inclusion in the top 10 list. While it isn't possible to determine which books would have been excluded from the LibraryReads list had the Hall of Fame authors been part of the normal list, it is possible to determine the number of books that would have been in the Hall of Fame had it been established earlier.

An alternate version of genre prediction using clusters of Library of Congress genre/form data rather than a single category may produce stronger results, however the relatively small sample size relative to the number of unique sets of genre/form classifiers may interfere with a closer match. It is also possible that the publisher may act as a strong genre signal, at least for fairly broad classifiers: for example, Avon is an imprint that focuses on publishing romance, and Tor focuses on science fiction and fantasy works. The publisher may not signal the genre at the same level of granularity as the Library of Congress genre/form headings, but for the broad categories, it may increase the accuracy of the results. 

Additional avenues of investigation that are not possible with the current dataset but could extend the analysis include using a larger set of catalog data to develop and train a classification model based on naturally occurring clusters, rather than the limited set of genres used in the LibraryReads Archive. Publishing history and sales data for authors featured in the Hall of Fame could act as a proxy for fanbase estimation, and help better understand the possible role of fanbase size in reaching the LibraryReads Hall of Fame. The IndieNext list is a similar book recommendation list, compiled from nominations by staff at independent bookstores. Examining overlap and differences between these lists may capture variations in trends for books readers purchase relative to books they want to obtain from their local library. 

## Limitations

The dominant limitation in executing analysis of this data is the limited sample size and relatively high dimensionality of the data, leading to a number of cases where a there is only one case of a particular classifier. There may be underlying trends in the titles that are not listed in the Library of Congress' Catalog. In the genre classifier model, absence of any Library of Congress data resulted in an assigment of Literary Fiction as the predicted genre. Data concerning publishers and inclusion or exclusion from the Library of Congress data is available in Appendix II.  

The broad genre categories limit my ability to look at granular trend changes within the broader genres or to identify genres blends.  

### Limitations in Interpretation and Generalization 

LibraryReads encourages publishers to advertise for the LibraryReads list as they distribute ARCs, both in hard copy and electronically. Success on the LibraryReads list may capture efficiacy in marketing as much as interest in the book itself. Publishers may also choose to limit the number of ARCs they distribute for a given title, or embargo the title entirely, which could prevent the title from reaching a sufficiently large audience to gain a spot on the list. Individuals may also vote entirely based on speculation, as they are not required to have read the book to cast their vote. LibraryReads is a vote based on popularity, and should not be interpreted as an indicator of literary merit, though the voting criteria is meant to capture desire to recommend the book to others. 

Publishers often provide the data that is found in the LC Catalog, and as noted above, Form/Genre choices may be part of a marketing decision as much as a classification of the content. While the form/genre list is a controlled vocabulary and criteria can be found on the Library of Congress' website (https://id.loc.gov/authorities/genreForms.html), actual application of the criteria is left to the cataloger. Using OCLC data may mitigate this, as a single work may have many classifications and genre may be derived from common classifications, but access and processing of OCLC data which may or may not use a controlled vocabulary has additional difficulties in access, scope, and cleaning.
